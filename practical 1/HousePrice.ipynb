```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data=pd.read_csv('/content/drive/MyDrive/ML_Lab/housepricedata.csv')
data
```

```python
data.info()
```

```python
x=data[["Area"]]
y=data["Price"]
```

```python
x
```

```python
y
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
area_to_predict = [[450]]
predicted_price = model.predict(area_to_predict)[0]
```

```python
r2 = r2_score(y, y_pred)
r2
```

```python
print(f"\nPredicted Price for Area = 450 sq ft: {predicted_price:,.2f}")
```

```python
plt.scatter(x, y, color='blue')            # Actual data points
plt.plot(x, model.predict(x), color='red') # Regression line
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression: Area vs Price')
plt.show()
```
